```python
from os import listdir
from os.path import isfile, join
import numpy as np
import networkx as nx
import operator

def input_graph(graph_path):
    fh = open(graph_path+".txt", "rb")
    GRAPH = nx.read_edgelist(fh)
    fh.close()
    nodes = GRAPH.nodes()
    #map = {n:int(i) for i, n in enumerate(nodes)}
    #GRAPH = nx.relabel_nodes(GRAPH, map)
    #GRAPH.remove_edges_from(nx.selfloop_edges(GRAPH))
    return GRAPH, map
def get_ci(g,l):
    ci = {}
    for i in g.nodes():
        ball = list(nx.single_source_shortest_path_length(g, i, cutoff=l))[0:]
        ball.remove(i)
        deg_ball = np.array([g.degree(b) for b in ball])        
        j = np.sum(deg_ball-1)
        ci[i] = (g.degree(i)-1)*j
    return dict( sorted(ci.items(), key=operator.itemgetter(1),reverse=True))

def get_list(graph_path,file):
    G,_ = input_graph(graph_path+file)
    action = []
    while len(G.nodes) != 0:
        ci = list(get_ci(G,3).keys())
        G.remove_node(ci[0])
        action.append(ci[0])
    with open(r'./'+file+".txt", 'w') as fp:
        for item in action:
            fp.write("%s\n" % item)
```
